#### Tipos de datos compuestos


Los datos compuestos en Python son tipos de datos que pueden almacenar colecciones de datos. Son fundamentales para manejar múltiples elementos de manera organizada y eficiente. Los principales tipos de datos compuestos en Python son listas, tuplas, diccionarios y conjuntos. Cada uno de estos tipos tiene características y usos específicos.

* **Listas**: Las listas son colecciones ordenadas y mutables que pueden contener elementos de diferentes tipos, incluidos otros compuestos de datos. Las listas son muy versátiles y se utilizan para almacenar secuencias de elementos que necesitan ser modificados, accedidos por índice, o iterados.

```python
nombres = ["Ana", "Luis", "Sofía"]
nombres.append("Carlos")  # Añadir un elemento al final
print(nombres[2])         # Acceder al tercer elemento, que es "Sofía"
```

* **Tuplas**: Las tuplas son colecciones ordenadas e inmutables. A diferencia de las listas, una vez que se crea una tupla, no se pueden modificar sus elementos. Esto las hace ideales para almacenar datos que no deben cambiar a lo largo del tiempo, como constantes o configuraciones.

```python
dimensiones = (1024, 768)
print(dimensiones[0])  # Acceder al primer elemento, que es 1024
```

* **Diccionarios**: Los diccionarios almacenan pares de clave-valor sin un orden específico. Las claves son únicas dentro de un mismo diccionario y son inmutables. Los diccionarios son mutables, lo que significa que se pueden añadir, modificar o eliminar pares clave-valor después de su creación. Son útiles cuando necesitas asociar pares de información y acceder a ellos rápidamente por clave.

```python
persona = {"nombre": "Marta", "edad": 28, "ciudad": "Madrid"}
print(persona["nombre"])  # Accede al valor que corresponde a la clave "nombre"
persona["edad"] = 29      # Modificar el valor asociado a la clave "edad"
```

* **Conjuntos**: Los conjuntos son colecciones no ordenadas de elementos únicos. Son mutables y útiles para eliminar duplicados de una colección, así como para realizar operaciones matemáticas de conjuntos, como la unión, intersección y diferencia.

```python
colores = {"rojo", "verde", "azul"}
colores.add("negro")         # Añadir un elemento
print("verde" in colores)    # Comprobar si "verde" está en el conjunto
```

```python
import nltk
nltk.download('punkt')
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.luhn import LuhnSummarizer

def summarize_with_luhn(text, language="spanish", summary_length=3):
    parser = PlaintextParser.from_string(text, Tokenizer(language))
    summarizer = LuhnSummarizer()
    summary = summarizer(parser.document, summary_length)
    return " ".join([str(sentence) for sentence in summary])

text = """
CONTRATO DE COMPRAVENTA

Entre los suscritos a saber, [Nombre del Vendedor], mayor de edad, domiciliado en [Ciudad, Estado], identificado con cédula de ciudadanía número [Número de Identificación], quien en adelante se denominará EL VENDEDOR, y por otra parte, [Nombre del Comprador], también mayor de edad, domiciliado en [Ciudad, Estado], identificado con cédula de ciudadanía número [Número de Identificación], que en adelante se denominará EL COMPRADOR, hemos acordado celebrar el presente contrato de compraventa sujeto a las siguientes cláusulas: \n

PRIMERA: OBJETO.- Por medio del presente contrato, EL VENDEDOR transfiere la propiedad a EL COMPRADOR del siguiente bien: [Descripción detallada del bien, incluyendo todas las especificaciones relevantes como modelo, marca, tamaño, color, número de serie, y cualquier otro dato que identifique plenamente el objeto de la venta].\n

SEGUNDA: PRECIO.- El precio acordado por las partes por la venta del bien descrito en la cláusula primera es de [Cantidad en números y letras], monto que EL COMPRADOR se obliga a pagar de la siguiente manera: [Describir el método de pago, como puede ser un pago total al momento de la firma, pagos en cuotas, etc., incluyendo fechas de pago o condiciones para realizarlos].\n

TERCERA: ENTREGA.- EL VENDEDOR se compromete a entregar el bien objeto de este contrato en [Dirección o lugar de entrega], en un estado que se corresponde con la descripción dada y libre de cargas y gravámenes al momento de la entrega.\n

CUARTA: GARANTÍAS Y DECLARACIONES.- EL VENDEDOR declara que tiene plena capacidad legal para vender el bien mencionado y asegura que dicho bien no cuenta con ningún tipo de limitación de disposición, cargas, gravámenes o impuestos pendientes.\n

QUINTA: RESPONSABILIDADES.- Cualquier desperfecto, obligación o responsabilidad que surja después de la entrega del bien será de cuenta y riesgo del COMPRADOR, salvo que se trate de una garantía explícitamente mencionada en este contrato.\n

SEXTA: JURISDICCIÓN.- Para todo lo relativo a la interpretación y cumplimiento del presente contrato, las partes se someten a las leyes y a los tribunales competentes de [Ciudad, Estado], renunciando a cualquier fuero que por razón de sus domicilios presentes o futuros pudiera corresponderles.\n

Firmado por las partes en [Ciudad], a los [Día] días del mes de [Mes] del año [Año].\n
"""

# Generate summary
summary = summarize_with_luhn(text)
print(summary)
```
