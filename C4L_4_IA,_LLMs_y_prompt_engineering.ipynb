### 2.2 ¿Qué es un LLM (modelo de lenguaje)?

Sin entrar en matemáticas: un **LLM** (*Large Language Model*) es un sistema que ha sido **entrenado con cantidades enormes de texto** (libros, artículos, web, código...) para hacer **una cosa muy concreta**:

> **Predecir qué texto viene después** de un texto dado.

Eso es. No "piensa". No "entiende" en el sentido humano. Lo que hace, simplificando mucho, es:

```
Le doy:   "Estimado cliente, le notificamos que..."
Devuelve: "...el día 15 de marzo tendrá lugar la audiencia..."
```

### Una analogía legal

Imagina un becario que ha **leído todos los contratos, sentencias y manuales jurídicos del mundo**, pero que:

- **No tiene memoria** de las conversaciones que tuvo ayer.
- **No sabe qué fue cierto** y qué fue inventado en lo que leyó.
- **Tiende a sonar seguro** incluso cuando se equivoca.

Eso es, a grandes rasgos, un LLM.

### Lo que un LLM SÍ hace bien

| Tarea | Por qué le va bien |
| --- | --- |
| **Resumir** texto largo | Entrenado con millones de resúmenes |
| **Traducir** entre idiomas | Visto muchísimo texto bilingüe |
| **Reformular** un texto en otro tono | Son patrones estadísticos claros |
| **Generar** código común | Hay muchísimo código público |
| **Estructurar** información en JSON, tabla, etc. | Los formatos están en su entrenamiento |

### Lo que un LLM NO hace bien (todavía)

| Limitación | Qué significa para un abogado |
| --- | --- |
| **Datos posteriores** a su fecha de corte | No conoce la jurisprudencia más reciente |
| **Cálculo numérico exacto** | Suma mal cuantías grandes; mejor delegar a Python |
| **Citas** | **Inventa referencias** que parecen reales — *alucinaciones* |
| **Casos muy específicos** | Si tu jurisdicción es local, puede tirar de patrones de otra |
| **Razonamiento jurídico delicado** | Sigue siendo cosa del letrado |

> **Un caso famoso (Mata vs. Avianca, EE.UU., 2023):** un abogado presentó un escrito ante un tribunal con seis citas jurisprudenciales que **no existían**. ChatGPT se las había inventado. Resultado: sanción al abogado.
>
> **Moraleja**: la IA es ayuda, no sustituto. Verifica SIEMPRE.

### 2.4 ¿Qué es un prompt?

Un **prompt** es **el texto que le damos al modelo** para que produzca una respuesta. Punto.

Pero no todos los prompts son iguales. Compara estos dos:

#### Prompt débil

```
Resume este contrato.
[contrato pegado]
```

#### Prompt fuerte

```
Eres un abogado especializado en derecho mercantil con 15 años de
experiencia en revisión de contratos.

Tarea: resume el contrato adjunto destacando:
  1. Las partes intervinientes.
  2. El objeto del contrato.
  3. Las cláusulas con riesgo legal alto (penalizaciones, exclusividad,
     indemnizaciones desproporcionadas, etc.).
  4. La duración y condiciones de terminación.

Formato: utiliza viñetas. Máximo 200 palabras. Responde en castellano.

Si detectas alguna cláusula que pueda contravenir la normativa española
de protección al consumidor (TRLGDCU), señálala con un AVISO al final.

[contrato pegado]
```

> ¿Ves la diferencia? El segundo prompt **dirige** al modelo. El primero le deja decidir.

### A esto se le llama "prompt engineering"

Y es una habilidad concreta y aprendible. Igual que sabes redactar un buen escrito procesal, puedes aprender a redactar un buen prompt.

Las tres reglas básicas:

1. **Cuanto más contexto y precisión, mejor**.
2. **Decid qué quieres Y en qué formato**.
3. **Si la primera respuesta no es buena, iterad** (no te rindas a la primera).


### 2.5 Los 4 ingredientes de un buen prompt

Un buen prompt suele tener **cuatro componentes**:

| # | Componente | Ejemplo |
| - | --- | --- |
| 1 | **Rol / contexto** | "Eres un abogado especialista en derecho laboral..." |
| 2 | **Tarea** | "Redacta una carta de despido por causas objetivas..." |
| 3 | **Formato de salida** | "Devuelve el texto en formato Markdown, con encabezados y máximo 300 palabras..." |
| 4 | **Ejemplos** *(opcional)* | "Ejemplo de tono deseado: [ejemplo previo]" |

### Plantilla mental

```
[ROL]
Eres un/a [especialidad] con experiencia en [contexto concreto].

[TAREA]
Necesito que [acción específica] sobre [el material que adjunto].
Concretamente, [paso 1, paso 2, paso 3...].

[FORMATO]
Devuelve la respuesta como [tipo de salida: tabla / viñetas / JSON / texto].
Longitud: [máximo X palabras / X líneas].
Idioma: [castellano de España / inglés...].
Tono: [formal / didáctico / coloquial].

[EJEMPLOS — opcional]
Por ejemplo, una respuesta válida sería: [...]

[MATERIAL]
[Aquí pegas el contrato, sentencia, hechos, etc.]
```

> **Truco práctico:** guárdate plantillas para tareas recurrentes (resumir contratos, buscar cláusulas abusivas, generar borradores de escritos...). Cada plantilla bien afinada te ahorra horas.


### 2.6 Patrones útiles (lo justo)

Te dejo cuatro patrones que cubren el 80% de los casos cotidianos. **El curso de pago entrará a fondo en muchos más** (chain-of-thought, few-shot estructurado, function calling, agents, etc.). Aquí, lo justo para empezar.

#### Patrón 1 — Asignar un rol

```
Eres un abogado especializado en X. [resto del prompt]
```

Ayuda al modelo a tomar el "tono" adecuado y a usar terminología correcta.

#### Patrón 2 — Pedir paso a paso

```
Antes de responder, razona el problema en pasos numerados.
Solo después de razonar, dame la respuesta final.
```

Mejora notablemente el razonamiento en problemas complejos. *(Lo veremos en profundidad en el curso de pago.)*

#### Patrón 3 — Pedir un formato concreto (especialmente JSON)

```
Devuélveme la respuesta como JSON con esta estructura:
{
  "partes": ["..."],
  "fecha": "AAAA-MM-DD",
  "cuantia": 0,
  "riesgo": "bajo|medio|alto"
}
No añadas nada más fuera del JSON.
```

Esto es **clave** cuando la salida del modelo va a ir a otro programa Python.

#### Patrón 4 — Permitir "no sé"

```
Si no tienes información suficiente para responder, di explícitamente
"No tengo información suficiente". No inventes.
```

Reduce considerablemente las alucinaciones.


### 2.7 Pitfalls — los errores que NO hay que cometer

#### 1) Confiar a ciegas

**El modelo se equivoca.** Y suena seguro al hacerlo. La regla de oro:

> Cualquier respuesta que vayas a usar **profesionalmente**, **VERIFICADLA** contra la fuente original.

#### 2) Dar datos sensibles a un modelo público

Si pegas un contrato con datos personales en ChatGPT consumer, esos datos **pueden ser usados** para entrenar futuros modelos (en algunos planes — lee la letra pequeña).

| Para datos sensibles, usa: | Por qué |
| --- | --- |
| **Modelos vía API con contrato profesional** | OpenAI / Anthropic API: no se entrena con tus datos. |
| **Despliegues privados** (Azure OpenAI, AWS Bedrock, etc.) | Datos dentro de tu entorno. |
| **Modelos abiertos en local** (Llama, Mistral) | Nada sale de tu red. |

> **NUNCA** pegues información de un cliente sin haber comprobado el plan que usas.

#### 3) Pedir "lo mejor" sin definir qué es mejor

> "Mejórame este contrato." → Pésimo prompt.

> "Reescribe la cláusula 5 para limitar la responsabilidad de mi cliente al importa pagado en los últimos 12 meses, manteniendo el resto del estilo del contrato." → Buen prompt.

#### 4) Usar la IA donde NO debe

| Tarea | ¿IA? |
| --- | --- |
| Resumen / traducción / redacción asistida | Sí (con verificación) |
| Búsqueda de jurisprudencia | **NO** sin verificar — alto riesgo de citas inventadas |
| Cálculos exactos (intereses, plazos) | **NO** — usa Python |
| Análisis de datos masivos | **Combinada con Python** — no a pelo |
| Tomar la decisión final del caso | **NUNCA** — es tu trabajo |


## 3. Python + IA: el equipo perfecto

Una pregunta natural: si la IA escribe código, ¿para qué aprendí Python?

### Porque la IA + Python = automatización a escala

Imagina esta tarea: **"Resumir 500 contratos del despacho y extraer los plazos de vencimiento."**

| Solo IA (manual) | IA + Python |
| --- | --- |
| Abrir cada contrato uno a uno | Cargar todos los archivos con Python |
| Pegar en ChatGPT | Llamar al modelo automáticamente |
| Copiar resumen | Guardar resumen en CSV/JSON |
| Repetir 500 veces (= 25 horas) | Esperar a que termina (= 30 minutos) |

### Cómo se ve esto en la práctica

A nivel **conceptual**, el flujo es:

```
[Carpeta con contratos]
        │
        ▼
   Python lee cada contrato
        │
        ▼
   Python envía el texto a la API del modelo (Claude, OpenAI...)
        │
        ▼
   Recibe el resumen
        │
        ▼
   Python guarda el resumen en una tabla
        │
        ▼
   Excel / CSV / informe PDF final
```

> **Importante:** los detalles técnicos (cómo se llama una API, autenticación, gestión de errores, costes, optimización...) los veremos en el **curso de pago**. Aquí solo queremos que tengas la **foto mental** correcta.

### Una pieza del rompecabezas — pseudocódigo

Sin entrar en detalles, esto es lo que se vería en un script real (no lo ejecutes — es ilustrativo):

```python
# Pseudocódigo — NO ejecutar (requiere API key, librerías específicas, etc.)

import os
from anthropic import Anthropic   # o "from openai import OpenAI"

client = Anthropic(api_key=os.environ["ANTHROPIC_API_KEY"])

prompt_base = '''
Eres un abogado mercantil. Resume el contrato adjunto destacando partes,
objeto, cuantía y duración. Devuelve la respuesta como JSON.
'''

for archivo in os.listdir("contratos"):
    with open(f"contratos/{archivo}") as f:
        texto = f.read()

    response = client.messages.create(
        model="claude-sonnet-4-6",
        max_tokens=1024,
        messages=[{"role": "user", "content": prompt_base + texto}]
    )

    print(archivo, "→", response.content[0].text)
```

> Si esto te pareció **legible**, eso significa que los tres días de Python han funcionado. ¡Enhorabuena!

### Ejemplo 1 — Resumen de un contrato

**Prompt débil:**
```
Resume este contrato.

[contrato]
```

**Prompt fuerte:**
```
Eres un abogado mercantil con experiencia en revisión de contratos
B2B en España.

Tarea: produce un resumen ejecutivo del contrato adjunto que incluye:

  1. Partes intervinientes (con su rol).
  2. Objeto del contrato.
  3. Cuantía o contraprestación económica.
  4. Duración y condiciones de prórroga / terminación.
  5. Cláusulas con riesgo elevado para la parte compradora:
     - Limitaciones de responsabilidad.
     - Exclusividad.
     - Penalizaciones desproporcionadas.
     - Cláusulas de jurisdicción exótica.
  6. Cualquier cláusula que pueda contravenir el TRLGDCU si una de las
     partes es consumidor.

Formato: viñetas, máximo 250 palabras, en castellano.

Si la información necesaria para algún apartado NO está en el texto,
indícalo expresamente con "No consta en el contrato".

[contrato]
```


### Ejemplo 2 — Comparar dos cláusulas

**Prompt débil:**
```
¿Cuál es mejor?

Cláusula A: [...]
Cláusula B: [...]
```

**Prompt fuerte:**
```
Eres un abogado especializado en arrendamientos de local de negocio.

Tarea: compara las dos cláusulas siguientes desde el punto de vista del
ARRENDATARIO. Para cada una, valora:

  1. Riesgo legal asumido por el arrendatario (1 = bajo, 5 = alto).
  2. Flexibilidad de salida.
  3. Coste potencial en caso de incumplimiento.
  4. Conformidad con la LAU vigente en España.

Devuelve la respuesta como una tabla Markdown con cuatro columnas
(criterio, cláusula A, cláusula B, recomendación), seguida de un
párrafo final con la cláusula más conveniente para el arrendatario
y por qué.

CLÁUSULA A:
[...]

CLÁUSULA B:
[...]
```


### Ejemplo 3 — Generar un primer borrador de escrito

**Prompt débil:**
```
Hazme una carta de reclamación para mi cliente.
```

**Prompt fuerte:**
```
Eres un abogado de consumo con experiencia en reclamaciones a
operadoras de telecomunicaciones en España.

Tarea: redacta una carta de reclamación previa de mi cliente
contra [OPERADORA] por los siguientes hechos:

  - Cliente: D./Dña. [NOMBRE], DNI [DNI], domiciliado en [DIRECCIÓN].
  - Servicio contratado: [tipo de tarifa] desde [fecha de alta].
  - Hechos: [descripción concreta — facturación incorrecta, baja
    no atendida, penalización indebida, etc.].
  - Pretensión: [reembolso de X€, baja efectiva, etc.].

Estructura del escrito:
  1. Encabezamiento (datos cliente y empresa).
  2. Antecedentes (hechos en orden cronológico).
  3. Fundamentos jurídicos (TRLGDCU, Ley General de
     Telecomunicaciones, normativa sectorial pertinente).
  4. Pretensión.
  5. Plazo de respuesta (15 días naturales).
  6. Reserva de acciones legales.

Formato: castellano formal, máximo 600 palabras, sin tecnicismos
innecesarios. Si te falta algún dato, márcalo entre [CORCHETES] para
que el abogado lo completa.
```

> Notad cómo en cada ejemplo "fuerte" hemos especificado: rol, tarea concreta, datos relevantes, estructura y formato. **Esa es toda la diferencia.**

## Glosario rápido

| Término | En palabras llanas |
| --- | --- |
| **IA generativa** | IA que crea texto, imágenes, código... a partir de una entrada. |
| **LLM (*Large Language Model*)** | Modelo entrenado para predecir y generar texto. Claude y ChatGPT son LLMs. |
| **Prompt** | El texto que damos al modelo para que produzca una respuesta. |
| **Prompt engineering** | El arte y la técnica de escribir buenos prompts. |
| **Token** | Cada "trozo" de texto que el modelo procesa (≈ 0.75 palabras de español). Los modelos cobran por token. |
| **Contexto** | El conjunto de información que le damos al modelo en un prompt. |
| **Ventana de contexto** | El máximo de tokens que un modelo puede procesar a la vez. |
| **Alucinación** | Cuando el modelo se inventa información que parece real pero no lo es. |
| **Temperatura** | Parámetro que controla la creatividad: 0 = muy determinista, 1 = más creativo. |
| **Modelo base / *foundation model*** | El modelo "puro", entrenado pero sin ajustar a una tarea concreta. |
| **API** | Interfaz para llamar al modelo desde código (Python, etc.). |
| **Sistema vs. usuario (en mensajes)** | "Sistema": instrucciones permanentes. "Usuario": lo que pides en cada turno. |
| **Function calling / tool use** | Mecanismo para que el modelo llame a funciones que tú defines. |
| **Agente** | Sistema con IA que decide y ejecuta acciones de forma semi-autónoma. |
| **RAG (*Retrieval-Augmented Generation*)** | Técnica para que el modelo use tus documentos como fuente. |
| **Fine-tuning** | Ajustar un modelo con tus datos para una tarea concreta (avanzado). |
| **Hallucination mitigation** | Técnicas para reducir invenciones (anti-alucinación). |
