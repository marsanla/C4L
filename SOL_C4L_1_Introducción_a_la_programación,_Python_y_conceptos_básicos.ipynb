## 3.2 Soluciones — tipos de datos y operaciones

Estos ejercicios introducen **operaciones aritméticas** y **fechas** (la sección 2.6 del notebook teórico explica `datetime` y `timedelta` a fondo).


### Ejercicio 1 — Calculadora de intereses de mora

**Idea:** interés = principal × tasa_anual × (días_retraso / 365).


```python
total = 2000              # Principal en euros
tasa_interes = 0.05       # 5%
dias_retraso = 90

interes_mora = total * tasa_interes * (dias_retraso / 365)
print(f"El interés de mora por {dias_retraso} días de retraso es: {interes_mora:.2f}€")

# Explicación:
# - Multiplicamos los tres valores en un solo paso.
# - Los paréntesis en (dias_retraso / 365) dejan claro el orden, aunque
#   matemáticamente Python lo resolvería igual.
# - {interes_mora:.2f} formatea el número con dos decimales.
```

### Ejercicio 2 — Conversor USD → EUR


```python
total = 1000            # dólares
tasa_cambio = 0.85      # 1 USD = 0.85 EUR

total_en_euros = total * tasa_cambio
print(f"{total}$ son equivalentes a {total_en_euros:.2f} €.")

# Explicación: una sola multiplicación. .2f para mostrar dos decimales.
```

### Ejercicio 3 — Alerta de vencimiento de contrato


```python
from datetime import datetime, timedelta

fecha_inicio = datetime.strptime("2024-01-01", "%Y-%m-%d")
duracion_dias = 365

fecha_vencimiento = fecha_inicio + timedelta(days=duracion_dias)
dias_restantes = (fecha_vencimiento - datetime.now()).days

print(f"Días restantes para el vencimiento del contrato: {dias_restantes}")

# Explicación:
# 1. strptime convierte el texto en un datetime.
# 2. fecha + timedelta(days=N) suma N días -> nuevo datetime.
# 3. fecha2 - fecha1 devuelve un timedelta; .days nos da el número de días.
# Si el contrato YA ha vencido, dias_restantes será negativo.
```

### Ejercicio 4 — Plazo de prescripción


```python
from datetime import datetime, timedelta

fecha_inicio = datetime.strptime("2020-01-01", "%Y-%m-%d")
plazo_prescripcion_anos = 5

fecha_prescripcion = fecha_inicio + timedelta(days=365 * plazo_prescripcion_anos)
print("Fecha de prescripción del derecho:", fecha_prescripcion.strftime("%Y-%m-%d"))

# Explicación:
# - timedelta no acepta "años" directamente.
# - Multiplicamos 365 * años como aproximación.
# - Para cálculos exactos usaríamos dateutil.relativedelta (curso avanzado).
```

### Ejercicio 6 — Resumen por palabras (slicing)


```python
descripcion_caso = '''
El demandado fue acusado de incumplimiento de contrato el 15 de marzo del 2024.
El caso involucra varias partes y está actualmente en proceso.
'''

# 1) Convertir el texto en lista de palabras (separadas por espacios)
palabras = descripcion_caso.split()

# 2) Tomar las 14 primeras con slicing [:14]
palabras_resumen = palabras[:14]

# 3) Volver a unirlas con un espacio entre cada una
resumen = " ".join(palabras_resumen)

print("Resumen del caso:", resumen)

# Explicación:
# - .split() sin argumentos divide el texto por espacios y saltos de línea.
# - [:14] toma los 14 primeros elementos de la lista (índices 0..13).
# - " ".join(...) hace la operación inversa: une la lista en un solo texto.
```

### Ejercicio 8 — Antigüedad laboral


```python
from datetime import datetime, timedelta

fecha_inicio = datetime.strptime("2015-04-22", "%Y-%m-%d")
antiguedad = datetime.now() - fecha_inicio

# datetime.now() - fecha_inicio devuelve un timedelta
# Convertimos los días totales a años, meses (aprox.) y días.
total_dias = antiguedad.days

años  = total_dias // 365
resto = total_dias %  365
meses = resto // 30
dias  = resto %  30

print(f"Antigüedad laboral: {años} años, {meses} meses y {dias} días.")

# Explicación:
# - // es la división entera (sin decimales).
# - %  es el resto de la división (lo que sobra).
# - Es una aproximación porque los meses no tienen exactamente 30 días.
```

### Ejercicio 9 — Generador de notificaciones


```python
def crear_notificacion(nombre, fecha, motivo):
    notificacion = (
        f"Estimado/a {nombre}, le notificamos que el día {fecha} "
        f"tendrá lugar {motivo}."
    )
    return notificacion


nombre_cliente = "Juan Pérez"
fecha_evento = "2024-04-22"
motivo_evento = "una audiencia preliminar sobre su caso"

notificacion_final = crear_notificacion(nombre_cliente, fecha_evento, motivo_evento)
print(notificacion_final)

# Explicación:
# - Definimos una función con `def`.
# - Recibe 3 parámetros y devuelve un texto formateado.
# - Las cadenas entre paréntesis se concatenan automáticamente — útil para
#   no tener líneas demasiado largas.
```

### Ejercicio 10 — Consulta de expediente anidado


```python
expedientes = {
    "001": {"nombre": "Carlos Smith", "tipo": "Divorcio", "estado": "Activo"},
    "002": {"nombre": "Ana Bolena",   "tipo": "Laboral",  "estado": "Cerrado"}
}

def mostrar_expediente(numero):
    expediente = expedientes[numero]   # accedemos al dict interno
    print(f"Expediente {numero}: {expediente}")


mostrar_expediente("001")

# Explicación:
# - `expedientes` es un diccionario donde cada VALOR es OTRO diccionario.
# - Accedemos al expediente concreto con expedientes["001"].
# - Si el número no existe, daría KeyError. Para evitar el error usaríamos:
#       expediente = expedientes.get(numero, "No encontrado")
```
